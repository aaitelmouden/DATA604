### Simulating Coin Toss Experiment

Let's toss a fair coin ten times "n=10" with the binomial distribution function, and repeat the experiment 1000 times, so
we can find out how many successes/heads we observed.

```python
import numpy as np
import random
import matplotlib.pyplot as plt

np.random.seed(42) # set the seed to get same random number
n=10
p=0.5
s=1000
x=np.random.binomial(n=n, p=p, size=s)
```

#### Probability of seeing x heads out of n=10 coin tosses

```python
[np.equal(x,i).sum() for i in range(n)]

# We can use the observed successes to estimate the probability of getting x successes in n=10 coin tosses by dividing by 1000.

probs_1000 = [np.equal(x,i).mean() for i in range(n)]

# Let us plot the probability of x successes that we just computed.

plt.xticks(range(n))
plt.plot(list(range(n)), probs_1000, color='blue', marker='o')
plt.xlabel('Number of Heads',fontsize=14)
plt.ylabel('Probability',fontsize=14)
plt.show()
```
